```python
# Caesar Cipher Encryption
def caesar_encrypt(text, shift):
    encrypted_text = ""  # Variable to store the encrypted result
    for char in text:  # Iterate through each character in the text to be encrypted
        if char.isalpha():  # Check if the character is a letter
            is_upper = char.isupper()  # Check if the character is uppercase
            char = char.lower()  # Convert the character to lowercase
            encrypted_char = chr(((ord(char) - 97 + shift) % 26) + 97)  # Encrypt the character using Caesar Cipher, then convert it back to lowercase
            if is_upper:  # Check if the original character was uppercase, then convert the encryption result to uppercase
                encrypted_char = encrypted_char.upper()
            encrypted_text += encrypted_char
        else:
            encrypted_text += char
    return encrypted_text

# Vigenere Cipher Encryption
def vigenere_encrypt(input_text, key):
    input_text = input_text.lower()  # Convert the text to lowercase
    key = key.lower()  # Convert the key to lowercase
    input_length = len(input_text)  # Calculate the length of the text
    key_length = len(key)  # Calculate the length of the key
    cipher_text = ''  # Variable to store the encrypted result
    j = 0  # Initialize the key index (starts from 0)

    for i in range(input_length):  # Iterate through each character in the text
        if input_text[i] == ' ':
            cipher_text += ' '  # If the character is a space, add a space to the encrypted text and continue to the next character
            continue
        if input_text[i] == ',':
            cipher_text += ','  # If the character is a comma, add a comma to the encrypted text and continue to the next character
            continue

        char_input = ord(input_text[i]) - 97  # Convert the character to a number from 0 to 25
        char_key = ord(key[j % key_length]) - 97  # Convert the corresponding key character to a number from 0 to 25
        char_cipher = (char_input + char_key) % 26  # Perform Vigenere encryption by calculating the modulo 26
        cipher_text += chr(char_cipher + 97)  # Convert the encryption result back to lowercase letters
        j += 1  # Move to the next key character (if it reaches the end of the key, wrap around to the beginning)

    return cipher_text  # Return the encrypted text

# Vigenere Cipher Decryption
def vigenere_decrypt(input_text, key):
    input_text = input_text.lower()  # Convert the text to lowercase
    key = key.lower()  # Convert the key to lowercase
    input_length = len(input_text)  # Calculate the length of the text
    key_length = len(key)  # Calculate the length of the key
    plain_text = ''  # Variable to store the decrypted result
    j = 0  # Initialize the key index

    for i in range(input_length):  # Iterate through each character in the text
        if input_text[i] == ' ':
            plain_text += ' '  # If the character is a space, add a space to the decrypted text and continue to the next character
            continue
        if input_text[i] == ',':
            plain_text += ','  # If the character is a comma, add a comma to the decrypted text and continue to the next character
            continue

        char_cipher = ord(input_text[i]) - 97  # Convert the encryption character to a number from 0 to 25
        char_key = ord(key[j % key_length]) - 97  # Convert the corresponding key character to a number from 0 to 25
        char_plain = (char_cipher - char_key + 26) % 26  # Perform Vigenere decryption
        plain_text += chr(char_plain + 97)  # Convert the decryption result back to lowercase letters
        j += 1  # Move to the next key character

    return plain_text  # Return the decrypted text

# Caesar Cipher Decryption
def caesar_decrypt(text, shift):
    decrypted_text = ""  # Variable to store the decrypted result
    for char in text:  # Iterate through each character in the text to be decrypted
        if char.isalpha():  # Check if the character is a letter
            is_upper = char.isupper()  # Check if the character is uppercase
            char = char.lower()  # Convert the character to lowercase
            decrypted_char = chr(((ord(char) - 97 - shift) % 26) + 97)  # Decrypt using the Caesar Cipher formula, then convert it back to lowercase
            if is_upper:  # Check if the original character was uppercase, then convert the decryption result to uppercase
                decrypted_char = decrypted_char.upper()
            decrypted_text += decrypted_char
        else:
            decrypted_text += char  # If it's not a letter, the decrypted character will remain unchanged
    return decrypted_text  # Return the decrypted text

# The text to be encrypted
original_text = "Success is not final, failure is not fatal, it is the courage to continue that counts"

# The key
key = "prima"

# Encrypt with Caesar Cipher
caesar_shift = 53
caesar_encrypted_text = caesar_encrypt(original_text, caesar_shift)

# Encrypt the Caesar Cipher encrypted result with Vigenere Cipher
vigenere_encrypted_text = vigenere_encrypt(caesar_encrypted_text, key)

# Decrypt the Vigenere Cipher encrypted result with Caesar Cipher
caesar_decrypted_text = caesar_decrypt(vigenere_encrypted_text, caesar_shift)

# Decrypt with Vigenere Cipher
vigenere_decrypted_text = vigenere_decrypt(caesar_decrypted_text, key)

# Output the encryption and decryption results
print("Original Message:", original_text)
print("Encrypted Message (Caesar):", caesar_encrypted_text)
print("Encrypted Message (Vigenere):", vigenere_encrypted_text)
print("Decrypted Message (Caesar):", caesar_decrypted_text)
print("Decrypted Message (Vigenere):", vigenere_decrypted_text)
```
